### 5.Keras :
Keras is an open-source library that provides a Python interface for artificial neural networks. Keras acts as an interface for the TensorFlow library. Up until version 2.3 Keras supported multiple backends, including TensorFlow, Microsoft Cognitive Toolkit, R, Theano, and PlaidML.

```python
import numpy as np
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt

url = "http://bit.ly/w-data"
df  = pd.read_csv(url)
df
```

## 2) Plot Graph of Hours vs Percentage 
to Understand the steps

```python
df.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

## 3) Assign Label and values 

```python
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values  
```

## 4) Creating a Model and Training the Model Using Sequential model of Keras 

### Sequential class of Keras:
Sequential groups a linear stack of layers into a tf.keras.Model.

Sequential provides training and inference features on this model.

### Optimizer :

In machine learning, Optimization is an important process which optimize the input weights by comparing the prediction and the loss function. Keras provides quite a few optimizer as a module, optimizers and they are as follows:

### SGD Class of Sequential :
Gradient descent (with momentum) optimizer.
SGD − Stochastic gradient descent optimizer.

### Dense Layers of Keras :
Dense implements the operation: output = activation(dot(input, kernel) + bias) where activation is the element-wise activation function passed as the activation argument, kernel is a weights matrix created by the layer, and bias is a bias vector created by the layer (only applicable if use_bias is True).
### Input shape of dense layer :

N-D tensor with shape: (batch_size, ..., input_dim). The most common situation would be a 2D input with shape (batch_size, input_dim).

### Losses :
The purpose of loss functions is to compute the quantity that a model should seek to minimize during training.
 ### mse comes under regression loss functions : 
 
###  Compiling the model :
Keras model provides a method, compile() to compile the model. The argument and default value of the compile() method is as follows

### Training the  Model :
Models are trained by NumPy arrays using fit(). The main purpose of this fit function is used to evaluate your model on training. This can be also used for graphing model performance. 
 

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])],[keras.layers.Dense(15)])
model.compile(optimizer='sgd', loss='mse')
model.fit(X, y, epochs=1500)
```

where ,

X, y− It is a tuple to evaluate your data.

epochs − no of times the model is needed to be evaluated during training.

batch_size − training instances.

## 5) Predicting the percentage can be scored by students who studies for 9.25 Hours

A class prediction is given the finalized model and one or more data instances, predict the class for the data instances.

We do not know the outcome classes for the new data. That is why we need the model in the first place.

```python
print(model.predict([9.25]))
```

#  ---:Analysis of accuracy of Prediction and error Percentage  (Extra):---

```python
list = df['Hours'].tolist()
```

```python
list1 = df['Scores'].tolist()
pred = []
m = []
```

### The Below set of codes illustrates the actual and predicted values in order to evaluate our prediction accuracy and error percentage  

```python
for i in range(len(list)):
    print("predicted", + model.predict([list[i]]),end = '    actual ')
    k = model.predict([list[i]])
    pred.append(k[0][0])
    m.append((k[0][0] - list[i])/list[i]) 
    print("  ", + list1[i])
```

```python
df['predicted'] = pred
df['error in percentage'] = m
```

### Below is the updated  data frame consist of : Hours,	Scores,	predicted  &	error in percentage 

```python
df
```

### Data visualisation for the above dataframe to understand the process of analysis result :

```python
df.plot(x='Hours', y=['Scores','predicted','error in percentage'], style=['*','o','+']) 
plt.title('Hours vs Percentage vs predicted vs error ')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```
